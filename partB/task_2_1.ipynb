# Task 2.1 Dataset Selection and Setup

## Dataset Justification
We synthesize a classification dataset using `sklearn.datasets.make_classification(n_samples=200, n_features=2, ...)`. This is a binary classification dataset. It is an appropriate testbed because the paper tests its proposed method on synthetic 2-dimensional Gaussian mixtures and checks binary classification under varying amounts of noise (Section: Experiments on Synthetic Data Sets). The main limitation compared to their dataset is that we only have two classes, but this makes the basic binary reproduction highly interpretable.

Following the paper's methodology, we generate our similarity matrix by taking the linear kernel $X X^T$ and adding Gaussian noise $N(0, \eta)$, directly yielding an indefinite kernel $K_0$.

```python
import numpy as np
from sklearn.datasets import make_classification
import os

np.random.seed(42)
X, y = make_classification(n_samples=200, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=42)
# Convert y to {+1, -1} for SVM
y = np.where(y == 0, -1, 1)

# Generate linear kernel
K_linear = X @ X.T

# Add additive white Gaussian noise N(0, eta) to make it indefinite
eta = 20.0
noise = np.random.normal(0, eta, K_linear.shape)
noise = (noise + noise.T) / 2
K_0 = K_linear + noise

os.makedirs("data", exist_ok=True)
np.savez("data/toy_dataset.npz", K_0=K_0, y=y, X=X)

print(f"Data shape: {X.shape}, y shape: {y.shape}")
print(f"Number of negative eigenvalues in K_0: {np.sum(np.linalg.eigvalsh(K_0) < 0)}")
```

The synthetic dataset with 200 samples and its indefinite similarity matrix $K_0$ are successfully created and stored.
